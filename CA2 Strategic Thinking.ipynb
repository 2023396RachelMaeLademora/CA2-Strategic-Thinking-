```python
# GridSearchCV to find optimal n_estimators
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# specify number of folds for k-fold CV
n_folds = 5

# parameters to build the model on
parameters = {'max_depth': range(1, 10, 2)}

# instantiate the model
clf = DecisionTreeClassifier()


# fit tree on training data
clf = GridSearchCV(clf, parameters, 
                    cv=n_folds, 
                   scoring="accuracy")
clf.fit(X_train, y_train)
```

I used this this because it will help to reduce the overfitting using the parameters. And the k folds is the number of the material that I need to test the score of the accuracy. 

```python
# scores of GridSearch CV
scores = clf.cv_results_
pd.DataFrame(scores).head()
```

I used this because I want to show and see the parametars of three materials that undergo replace. 

```python
# plotting accuracies with max_depth
plt.figure()
plt.plot(scores["param_max_depth"], 
         scores["mean_test_score"], 
         label="test accuracy")
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

I used this to properly vizualize the plot of the parameter max depth because as I increase the max depth as shown in the graph it increase during the test score  until at the variance of 1 and then it goes down to the variance of 0.95. Therefore, this is very helful to reduce the overfitting of the decision tree accuracy. 

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [6,9,10],
    'min_samples_leaf': range(300, 500, 500),
    'min_samples_split': range(400, 700, 800),
    'n_estimators': [300,400, 500], 
    'max_features': [5, 10]
}
# Create a based model
clf = DecisionTreeClassifier()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = clf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1,verbose = 1)
```

I used this to as a parameter grid as to prepare the model for testing in different parameters and I put the clf decision tree before the initiating the grid search model. This is vital beacsue this will be the parameter value test numbersto train the model. 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
# Create an instance of the DecisionTreeClassifier
estimator = DecisionTreeClassifier()

# Define the parameter grid without 'max_depth'
param_grid = {}

# Create an instance of GridSearchCV with the estimator and parameter grid
grid_search = GridSearchCV(estimator, param_grid, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

I used this to estimate the decision tree classifier with the estimator parameter to properly implement the accurate accuracy test score using the grid search CV. This will help to fit the X train and y train testing of accuracy.
