### One Hot Coding 

```python
# from discrption of columns data there are some columns don't needed so I drop them
df_new = df.copy()
df_new.drop(columns=['Precursor', 'Activation agent', 'Reference', 'Unnamed: 12', 'Unnamed: 13'], inplace=True)
df_new.head()
```

```python
df_new = pd.get_dummies(df_new)
df_new.head()
```

```python
df_new.drop(columns= 'Adsorption pressure_bar', inplace=True)
```

```python
df_new.head()
```

### Split the data

```python
X = df_new.drop(columns='Material_replace')
y = df_new.Material_replace
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Impute Data

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn import impute
```

```python
X_train.isna().sum()
```

```python
imputer = impute.IterativeImputer()
num_cols = ['Activation temperature_°C', 'BET surface area_m2/g', 'Total pore volume_cm3/g', 'CO2 uptake_mmol/g_25°C']
imputed = imputer.fit_transform(X_train[num_cols])
X_train.loc[:,num_cols] = imputed
```

```python
X_train.isna().sum()
```

```python
X_train.isna().sum()
```

```python
X_test.isna().sum()
```

```python
imputed = imputer.fit_transform(X_test[num_cols])
X_test.loc[:, num_cols] = imputed
```

```python
X_test.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.dummy import DummyClassifier
bm = DummyClassifier()
bm.fit(X_train, y_train)
bm.score(X_train, y_train)
```

```python
bm.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=42,
                            max_depth=3)

clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
#pip install yellowbrick
```

```python
from yellowbrick.classifier import (ConfusionMatrix, ROCAUC, ClassificationReport, 
ClassPredictionError, ClassBalance)
```

```python
mapping = {'Activated Carbon': 0, 'Metal-organic frameworks (MOFs': 1, 'Microporous polymers': 2}

actual_labels = ['Activated Carbon', 'Metal-organic frameworks (MOFs', 'Microporous polymers']

cm_viz = ConfusionMatrix(clf, 
                         classes=actual_labels, label_encoder=mapping)

cm_viz.score(X_test, y_test)
```

```python
actual_labels = ['Activated Carbon', 'Metal-organic frameworks (MOFs', 'Microporous polymers']
fig, ax = plt.subplots(figsize=(6,4))
cm_viz = ClassificationReport(clf,
                        classes=['Activated Carbon', 'Metal-organic frameworks (MOFs', 'Microporous polymers'],
                        label_encoder=mapping)
cm_viz.score(X_test, y_test)
cm_viz.poof()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
cm_viz = ClassPredictionError(clf,
                        classes=['Activated Carbon', 'Metal-organic frameworks (MOFs', 'Microporous polymers'])
cm_viz.score(X_test, y_test)
cm_viz.poof()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
cb_viz = ClassBalance(labels=['Activated Carbon', 'Metal-organic frameworks (MOFs', 'Microporous polymers'])

cb_viz.fit(y)
cb_viz.poof()
```

```python
viz = ROCAUC(clf, classes=['Activated Carbon', 'Metal-organic frameworks (MOFs', 'Microporous polymers'])

viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.show()
```
